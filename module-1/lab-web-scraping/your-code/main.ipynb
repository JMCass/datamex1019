#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url11 = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code
html11=requests.get(url11).content
html11[0:800]
````

````python
soup11 = bs(html11, 'html.parser')
soup11
````

````python
idiomas=soup11.find('table')
idiomas
````

````python
lst_lan=[]
for e in idiomas.find_all('tr'):
    fila=[f for f in e.find_all('td')]
    if len(fila)>0:
        lst={'Language':fila[1].find('a').text.strip(),
              'Country':fila[2].find('a').text.strip(),
              'Speakers (millions)':fila[4].text.replace('\n','')}
        lst_lan.append(lst)
````

````python
lst_lan[0:11]
````

````python
import pandas as pd
````

````python
df=pd.DataFrame(lst_lan)
df
````
