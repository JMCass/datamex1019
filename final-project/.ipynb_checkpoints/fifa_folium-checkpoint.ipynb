```python
import pandas as pd
import folium
import numpy as np
import matplotlib.pyplot as plt
```

```python
geoloc_club=pd.read_csv('./Club.csv', index_col=0)
geoloc_club.head()
```

```python
#continent = geoloc_club.groupby(['Latitude', 'Longitude', 'Stadium'])['Continent']
```

```python
map_club=folium.Map(location=[0.0, 0.0], tiles='openstreetmap', zoom_start=1)
```

```python
lat = geoloc_club['Latitude']
lng = geoloc_club['Longitude']
label = geoloc_club['Stadium']
```

```python
for i in range(len(lat)):
    folium.Marker([lat[i], lng[i]], popup=folium.Popup('Name: {}'.format(label[i]), parse_html=True, max_width=80)
                  ,icon=folium.Icon(icon='cloud')).add_to(map_club)
```
