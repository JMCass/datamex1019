```python
import pandas as pd
import folium
import numpy as np
import matplotlib.pyplot as plt
```

```python
geoloc_club=pd.read_csv('./Club.csv', index_col=0)
geoloc_club.tail()
```

```python
america_df = geoloc_club['Continent'] == 'America'
df1 = geoloc_club[america_df].reset_index()
```

```python
ame_map=folium.Map(location=[0.0, 0.0], tiles='openstreetmap', zoom_start=2)
```

```python
lat1 = df1['Latitude']
lng1 = df1['Longitude']
label1 = df1['Stadium']
```

```python
for i in range(len(lat1)):
    folium.Marker([lat1[i], lng1[i]], popup=folium.Popup('Name: {}'.format(label1[i]), parse_html=True, max_width=80)
                  ,icon=folium.Icon(icon='cloud')).add_to(ame_map)
```
