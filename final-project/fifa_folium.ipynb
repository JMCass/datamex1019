```python
import pandas as pd
import folium
import numpy as np
import matplotlib.pyplot as plt
```

```python
geoloc_club=pd.read_csv('./Club.csv', index_col=0)
geoloc_club.head()
```

```python
america_df = geoloc_club['Continent'] == 'America'
df1 = geoloc_club[america_df].reset_index()
```

```python
ame_map=folium.Map(location=[19.302778, -99.150556], tiles='cartodbpositron', zoom_start=2)
```

```python
lat1 = df1['Latitude']
lng1 = df1['Longitude']
label1 = df1['Stadium']
country1 = df1['Nationality']
division1 = df1['Div']
club1 = df1['Team']
```

```python
for i in range(len(lat1)):
    folium.CircleMarker([lat1[i], lng1[i]], radius=1, color='red', popup=('<b>''Club: ''</b>' + club1[i] + '<br>'+
                                                                          '<b>''Stadium: ''</b>' + label1[i] + '<br>'+
                                                                          '<b>''Country: ''</b>' + country1[i] + '<br>'+ 
                                                                          '<b>''Division: ''</b>' + division1[i])
                  ,icon=folium.Icon()).add_to(ame_map)
```

```python
ame_map
```

```python
ame_map.save('america.html')
```
