## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
unique_amount = orders.groupby('CustomerID').agg({'amount_spent':'sum'}).sort_values('amount_spent', ascending=False)

unique_amount
```

```python
q95 = float(unique_amount.quantile(0.95))
q95
```

```python
q75 = float(unique_amount.quantile(0.75))
q75
```

```python
vip = unique_amount[unique_amount.amount_spent>q95]
vip
```

```python
prefered = unique_amount[(unique_amount.amount_spent<=q95) & (unique_amount.amount_spent>q75)]
prefered
```

```python
def label(x):
    if x in vip.index:
        x = 'VIP'
    elif x in prefered.index:
        x = 'Preferred'
    else: 
        x = 'New'
    return x
```

```python
orders = orders.assign(labels=orders.CustomerID)
```

```python
orders.labels = orders.labels.apply(label)
orders
```

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# your code here for Q2

get_vip = orders[orders.labels == 'VIP']

vip_countries = get_vip.drop_duplicates(subset=['CustomerID','Country'])
vip_countries
```

```python
lst_vip = vip_countries.groupby('Country').agg({'CustomerID': 'count'}).sort_values('CustomerID',ascending=False)
lst_vip
```

```python
# your code here
get_pref = orders[(orders.labels == 'Preferred')|(orders.labels == 'VIP')]

pref_countries = get_pref.drop_duplicates(subset=['CustomerID','Country'])
pref_countries
```

```python
lst_vip_pref = pref_countries.groupby('Country').agg({'CustomerID': 'count'}).sort_values('CustomerID',ascending=False)
lst_vip_pref
```
